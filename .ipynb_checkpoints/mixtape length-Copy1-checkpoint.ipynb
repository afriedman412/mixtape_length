```python
import pandas as pd
pd.set_option('display.precision',3)

import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.patches as mpatches
%matplotlib inline
plt.style.use("ggplot")

import numpy as np
import requests
import urllib
import urllib3
import certifi
import json
import re
import unidecode
import datetime as datetime
from apiclient.discovery import build
from apiclient.errors import HttpError
from youtube import API
from sklearn.linear_model import LinearRegression
```

```python
# instantiate the API interface
DEVELOPER_KEY = 'your key goes here'
YOUTUBE_API_SERVICE_NAME = "youtube"
YOUTUBE_API_VERSION = "v3"

youtube = build(YOUTUBE_API_SERVICE_NAME, YOUTUBE_API_VERSION,developerKey=DEVELOPER_KEY)
```

```python
# some functions for importing data
def normalize_column(c):
    return c.map(lambda x:(x-min(c))/(max(c)-min(c)))

def playlist_plays(playlist, less_than_first=False, artist=None, project=None):
    
    # pull raw data
    data_raw = youtube.playlistItems().list(
        part='contentDetails',
        playlistId=playlist,
        maxResults='50'
    ).execute()
    
    # initiate catalog
    pl_data = {'track':[], 'id':[], 'plays':[]}
    
    # iterate through raw data and pull entries
    for v in data_raw['items']:
        stat = youtube.videos().list(
            id=v['contentDetails']['videoId'], part='statistics, snippet').execute()
        pl_data['track'].append(stat['items'][0]['snippet']['title'])
        pl_data['id'].append(v['contentDetails']['videoId'])
        pl_data['plays'].append(int(stat['items'][0]['statistics']['viewCount']))
        
    
    # normalize plays
    pl_data['norm_plays'] = []
    for p in pl_data['plays']:
        pl_data['norm_plays'].append((p-min(pl_data['plays']))/(max(pl_data['plays'])-min(pl_data['plays'])))

    
    # create dataframe
    pl_df = pd.DataFrame(pl_data)
    
    # add date feature
    pl_df['date'] = datetime.datetime.today().strftime('%Y-%m-%d')
    
    # add project info
    pl_df['artist'] = artist
    pl_df['project'] = project
    
    # convert index to track position feature
    pl_df = pl_df.reset_index()
    pl_df = pl_df.rename(columns={'index':'position'})
    pl_df['length'] = len(pl_df)
        
    # remove tracks with more plays than first track
    if less_than_first is True:
        
        # define first track plays and length of project
        first = pl_df.loc[0,'plays']
        og_length = len(pl_df)
        
        # remove tracks with more plays than first track
        pl_df = pl_df[~((pl_df.position != 0) & (pl_df.plays > first))]
        
        # return nothing if less than %60 of tracks come back
        if len(pl_df)/og_length < 0.6:
            print('less than 0.6')
            return
        
        # re-normalize plays
        pl_df['renorm_plays'] = normalize_column(pl_df['plays'])
    
    return pl_df
```

Load in external file with mixtape links.

Create full dataset using mixtape links and `playlist_plays` code.

```python
out_df = pd.DataFrame()
for r in link_df.iterrows():
    print(r[1]['artist'], r[1]['album'])
    try:
        p_df = playlist_plays(r[1]['id'], artist=r[1]['artist'], project=r[1]['album'], less_than_first=True)
    except:
        print('error!')
        continue
    print('')
    out_df = out_df.append(p_df)
out_df['plays'] = out_df['plays'].map(int)
```
